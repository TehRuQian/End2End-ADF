```python
configs = {
    "fs.azure.account.auth.type": "CustomAccessToken",
    "fs.azure.account.custom.token.provider.class": spark.conf.get("spark.databricks.passthrough.adls.gen2.tokenProviderClassName")
}

#Optionally, you can add <directory-name> to the source URI of your mount point.
df = spark.read.format("parquet").load(
    "abfss://bronze@intechsacc.dfs.core.windows.net/SalesLT / Address/Address.parquet"
)
```

```python
dbutils.fs.ls("abfss://bronze@intechsacc.dfs.core.windows.net/SalesLT / Address/")
```

```python
df.write.mode("overwrite").parquet(
    "abfss://silver@intechsacc.dfs.core.windows.net/"
)

df.write.mode("overwrite").parquet(
    "abfss://gold@intechsacc.dfs.core.windows.net/"
)
```

dbutils.fs.ls("/mnt/bronze/SalesLT")

dbutils.fs.mount(
  source = "abfss://silver@intechsacc.dfs.core.windows.net/",
  mount_point = "/mnt/silver",
  extra_configs = configs)
